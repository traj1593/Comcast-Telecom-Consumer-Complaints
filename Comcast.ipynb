```python
df.groupby(["State"]).size().sort_values(ascending=False).to_frame().rename({0: "Count"},axis=1)
```

```python
status_complaints = df.groupby(["State","newStatus"]).size().unstack().fillna(0)
status_complaints
```

```python
status_complaints.plot(kind="barh",figsize=(20,30),stacked=True)
plt.rcParams.update({"font.size":30})
```

```python
max_complaint = df.groupby(["State"]).size().sort_values(ascending=False).to_frame().rename({0: "Count"},axis=1)[:1]
max_complaint.plot(kind="barh")
plt.rcParams.update({"font.size":30})
```

```python
unresolved_complaint_percentage = (status_complaints["Open"]/(status_complaints["Open"]+status_complaints["Closed"]))*100
unresolved_complaint_percentage.plot(kind="barh",figsize=(20,30),stacked=True)
plt.rcParams.update({"font.size":30})
```

```python
unresolved_complaint_percentage = (status_complaints["Closed"]/(status_complaints["Open"]+status_complaints["Closed"]))*100
unresolved_complaint_percentage.plot(kind="barh",figsize=(20,30),stacked=True)
plt.rcParams.update({"font.size":30})
```
